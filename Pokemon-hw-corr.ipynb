```python
import pandas as pd
pokemon_df = pd.read_csv('pokemon.csv', low_memory=False)
```

```python
# Find out the correlation between heigh/weigth with othe base stats
pokemon_df[["height_m", "weight_kg", "hp", "attack", "defense", "sp_attack", "sp_defense","speed"]].corr()
```

```python
# plot a scatter to identify the data trends for Height Vs Various Base Stats
pokemon_df.plot(
    kind='scatter',
    x='height_m', 
    y=['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed'], 
    backend='plotly', 
    width=700,
    title="Height Vs Various Base Stats"
)
```

```python
# plot a scatter to identify the data trends for Weight Vs Various Base Stats
pokemon_df.plot(
    kind='scatter',
    x='weight_kg', 
    y=['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed'], 
    backend='plotly', 
    width=700, 
    title="Weight Vs Various Base Stats"  
)
```
